### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
t1=set(map(int,input("Enter a set A:").split()))
t2=set(map(int,input("Enter a set B:").split()))

print("Union :",t1.union(t2))
print("Intersection :",t1.intersection(t2))
print("Difference(A-B) :",t1.difference(t2))
````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
keys = list(input("Enter keys for dict :").split())
values = list(map(int,input("Enter value for dict :").split()))
dict1 = {k: v for k, v in zip(keys, values)}

keys = list(input("Enter keys for dict :").split())
values = list(map(int,input("Enter value for dict :").split()))
dict2 = {k: v for k, v in zip(keys, values)}

result = { k : dict1[k] for k in dict1.keys()}
for key, value in dict2.items():
    if key in result:
        result[key] += int(value)
    else:
        result[key] = int(value)
print(result)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
keys = list(input("Enter keys for dict :").split())
values = list(map(int,input("Enter value for dict :").split()))
dict1 = {k: v for k, v in zip(keys, values)}
print("original dictionary :", dict1)
dict1 = dict( (sorted(dict1.items())))
print("Sorted dictionary :", dict1)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
keys = list(input("Enter keys for dict :").split())
values = list(map(int,input("Enter value for dict :").split()))
dict1 = {k: v for k, v in zip(keys, values)}

max_value = max(dict1.values()) 
max_key = [k for k, v in dict1.items() if v == max_value]
min_value = min(dict1.values())
min_key = [k for k, v in dict1.items() if v == min_value]

print(f"The maximum key is: {max_key}")
print(f"The minimum key is: {min_key}")
````

### 08) WAP to create a dictionary with numbers from 1 to N as keys and their squares as values.

**Sample Input:**
```
N = 5
```
**Sample Output:**
```
{1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

````python
number = int(input("Enter value of N:"))
dict1=dict()
for i in range(1,number):
    dict1[i] = (i*i)
print(dict1)
````

### 09) WAP to count the frequency of each character in a string using a dictionary.
**Sample Input:**
```
programming
```
**Sample Output:**
```
{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 1, 'n': 1}
```

````python
word = input("Enter a word:")
dict1=dict()
for i in word:
    if i in dict1 :
        dict1[i] += 1
    else:
        dict1[i] = 1
print(dict1)
````

### 10) WAP to swap keys and values of a dictionary.

