### 01) WAP to check whether the given number is positive or negative.

```python
number = int(input("Enter a Number : "))
if number > 0:
    print(f"{number} is positive")
elif number== 0:
    print(f"{number} is zero")
else:
    print(f"{number} is negative")
```

### 03) WAP to find out largest number from given two numbers using simple if and ternary operator.

```python
num1 , num2 =input("Enter two number with space :").split(" ")
num1=int(num1)
num2=int(num2)
print("using simple if operator:")
if num1 > num2:
    print(f"{num1} is greater than {num2}")
elif num1 == num2 :
    print(f"{num1} is equal to {num2}")
else:
    print(f"{num2} is greater than {num1}")

print("using ternary operator:")

print(f"{num1} is greater than {num2}") if num1 > num2 else print(f"{num1} is equal to {num2}") if num1 == num2 else print(f"{num2} is greater than {num1}")
```

### 04) WAP to find out largest number from given three numbers.

```python
num1 , num2 , num3 =input("Enter two number with space :").split(" ")
num1=int(num1)
num2=int(num2)
num3=int(num3)
print("using simple if operator:")
if num1 > num2 and num1 > num3:
    print(f"{num1} is greater")
elif num2 > num3 :
    print(f"{num2} is greater")
else:
    print(f"{num3} is greater")

print("using ternary operator:")

print(f"{num1} is greater") if num1 > num2 and num1 > num3 else print(f"{num2} is greater") if num2 > num3 else print(f"{num3} is greater")
```

### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter a Year :"))
if (year%400==0) or (year%4==0 and year%100!=0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
num1 , num2 =input("Enter two number with space :").split(" ")
operator = input("Enter an Operator(+,-,*,/)")
num1=int(num1)
num2=int(num2)
match operator:
    case "+":
        print(f"Addition of {num1} and {num2} = {num1+num2}")
    case "-":
        print(f"Subtraction of {num1} and {num2} = {num1-num2}")
    case "*":
        print(f"Multiplication of {num1} and {num2} = {num1*num2}")
    case "/":
        print(f"Division of {num1} and {num2} = {num1/num2}")
    case _:
        print("Invalid Operator")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
m1 , m2 , m3 , m4 , m5 = input("Enter marks of 5 subject with space :").split(" ")
m1=float(m1)
m2=float(m2)
m3=float(m3)
m4=float(m4)
m5=float(m5)
percentage = float( ( (m1+m2+m3+m4+m5) /500 ) *100 )
if percentage < 35:
    print(f"Percentage : {percentage} , Fail")
elif percentage >= 35 and percentage < 45 :
    print(f"Percentage : {percentage} , Pass")
elif percentage >= 45 and percentage < 60 :
    print(f"Percentage : {percentage} , Second")
elif percentage >= 60 and percentage < 70:
    print(f"Percentage : {percentage} , First Class")
else :
    print(f"Percentage : {percentage} , Distinction")
```
