### 01) WAP to find sum of all the elements in a List.

```python
length = int(input("Enter number of elements: "))
numbers = []

for i in range(length):
    num = int(input(f"Enter element {i+1}: "))
    numbers.append(num)

total = sum(numbers)

print("List elements:", numbers)
print("Sum of all elements:", total)
```

### 02) WAP to find largest element in a List.

```python
length = int(input("Enter number of elements: "))
numbers = []

for i in range(length):
    num = int(input(f"Enter element {i+1}: "))
    numbers.append(num)

max = max(numbers)

print("List elements :" , numbers)
print("Maximum element :" ,max )
```

### 04) WAP to interchange first and last elements in a list.

### 05) WAP to split the List into two parts and append the first part to the end.

```python
length = int(input("Enter number of elements: "))
numbers = []

for i in range(length):
    numbers.append(int(input(f"Enter element {i+1}: ")))

position = int(input("Enter position to split the list: "))

result = numbers[position:] + numbers[:position]

print("Original List:", numbers)
print("Modified List:", result)
```

### 06) WAP to interchange the elements on two positions entered by a user.

```python
length = int(input("Enter number of elements: "))
numbers = []

for i in range(length):
    numbers.append(int(input(f"Enter element {i+1}: ")))

position1 , position2 = input("Enter position of elements: ").split(" ")
position1 = int(position1) - 1
position2 = int(position2) -1
print("list elements :" , numbers)

numbers[position1] , numbers[position2] = numbers[position2] , numbers[position1]

print("New list elements :" , numbers)
```

### 07) WAP to reverse the list entered by user.

```python
length = int(input("Enter number of elements: "))
numbers = []

for i in range(length):
    numbers.append(int(input(f"Enter element {i+1}: ")))

print("Original List:", numbers)
numbers.reverse()
print("Reversed List:", numbers)
```

### 08) WAP to print even numbers in a list.

```python
numbers = list(map(int , input("Enter numbers with space:").split(" ")))
result = [i for i in numbers if i%2==0]
print("Original List:", numbers)
print("Even numbers :",result)
```

### 09) WAP to count unique items in a list.

```python
numbers = list(map(int , input("Enter numbers with space:").split(" ")))
count1=0
for i in numbers:
    if numbers.count(i) == 1:
        count1+=1
print("Original List:", numbers)
print("total Unique items:", count1)
```

### 11) WAP to print all odd numbers in a given range.

```python
start , end = input("Enter Starting and Ending Numbers :").split()
start = int(start)
start = start+1 if start%2==0 else start
end = int(end)

numbers= [ i for i in range(start ,end ,2)]
print("Odd numbers :",numbers)
```

### 12) WAP to count occurrences of an element in a list.

```python
numbers = list(map(int , input("Enter numbers with space:").split(" ")))
count = []
for i in numbers:
    if i not in count:
        print("Count of",i,"=",numbers.count(i))
        count.append(i)
```

### 14) WAP to extract elements with frequency greater than K.

```python
numbers = list(map(int , input("Enter numbers with space:").split(" ")))
k= int(input("Enter Frequency :"))
res = []
for i in numbers:
    if numbers.count(i) > k:
            res.append(i)

print("Original List:", numbers)
print("frequency greater than K items:",list(set(res)))
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
fruits = input("Enter numbers with space:").split(" ")
res=[]
for i in fruits:
    if i[0]  in "bB":
        res.append(i)
print(res)
```

### 17) WAP to create a list of common elements from given two lists.

```python
numbers1 = list(map(int , input("Enter numbers with space:").split(" ")))
numbers2 = list(map(int , input("Enter numbers with space:").split(" ")))
common = [i for i in numbers1 if i in numbers2]

print(common)
```
