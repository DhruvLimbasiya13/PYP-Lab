### %matplotlib inline — Inline plotting

```python
%matplotlib inline
import matplotlib.pyplot as plt

categories = ['Engineering', 'Design', 'Marketing', 'Sales', 'Support']
slices = [40, 20, 15, 15, 10]
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#FFC300']
explode = [0.05, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize=(4, 4))
wedges, texts, autotexts = ax.pie(
    slices, 
    labels=categories, 
    colors=colors, 
    autopct='%1.1f%%', 
    startangle=90, 
    explode=explode, 
    pctdistance=0.85,
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
ax.axis('equal')
plt.title('Department Distribution', fontsize=16, fontweight='bold')
plt.show()
```

### %%time — Time the entire cell (once)
#### Measures execution time of the whole cell. Runs it once.

```python
%%time
total = 0
for i in range(10):
    total += i
print("Total is", total)
# CPU times: total: 15.6 ms
# Wall time: 13.4 ms
```

```python
%%writefile file1.py
total = 0
for i in range(3):
    total += i
print("Total is", total)
# Writing file1.py
```

```python
import timeit
time_list_comp = timeit.timeit('[x**2 for x in range(100)]', number=1000000)
time_map = timeit.timeit('list(map(lambda x: x**2, range(100)))', number=1000000)

print(f"List comprehension time: {time_list_comp} seconds")
print(f"Map function time: {time_map} seconds")

if time_list_comp < time_map:
    print("List comprehension is faster.")
else:
    print("Map function is faster.")
```
