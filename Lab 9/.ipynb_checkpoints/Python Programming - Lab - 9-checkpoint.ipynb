### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def bmi(mass, height):
    return mass / (height ** 2)

m = float(input("Enter mass (kg): "))
h = float(input("Enter height (m): "))
print(bmi(m, h))
```

### 02) Write a function that add first n numbers.

```python
def sum_n(n):
    return n * (n + 1) // 2

n = int(input("Enter n: "))
print(sum_n(n))
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def is_prime(n):
    if n < 2:
        return 0
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return 0
    return 1

n = int(input("Enter number: "))
print(is_prime(n))
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primes_between(a, b):
    primes = []
    for n in range(a, b + 1):
        if n > 1:
            for i in range(2, int(n**0.5) + 1):
                if n % i == 0:
                    break
            else:
                primes.append(n)
    return primes

a = int(input("Enter start: "))
b = int(input("Enter end: "))

result = primes_between(a, b)
print(result)
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def is_palindrome(s):
    return s == s[::-1]

s = input("Enter string: ")
print(is_palindrome(s))
```

### 06) Write a function that returns the sum of all the elements of the list.

```python
def list_sum(lst):
    return sum(lst)

lst = list(map(int, input("Enter list elements: ").split()))
print("Sum of List :",list_sum(lst))
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sum_first_elements(tuples):
    return sum(t[0] for t in tuples)

n = int(input("Enter number of tuples: "))
tuples = []

for _ in range(n):
    t = tuple(map(int, input("Enter tuple: ").split()))
    tuples.append(t)

print(sum_first_elements(tuples))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

n = int(input("Enter n: "))
print(fib(n))
```

### 09) Write a function to get the name of the student based on the given rollno.
#### Example: Given dict1 = {101:'Ajay', 102:'Rahul', 103:'Jay', 104:'Pooja'} find name of student whose rollno = 103

```python
def get_student(roll, d):
    if roll in d:
        return d[roll]
    else:
        return "Roll number not found"

n = int(input("Enter number of students: "))
students = {}

for _ in range(n):
    r = int(input("Roll: "))
    name = input("Name: ")
    students[r] = name

roll = int(input("Enter roll to search: "))
print(get_student(roll, students))
```
