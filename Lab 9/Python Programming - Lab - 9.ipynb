### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def is_prime(n):
    if n < 2:
        return 0
    for i in range(2, int(n//2) + 1):
        if n % i == 0:
            return 0
    return 1

n = int(input("Enter number: "))
print(is_prime(n))
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primes_between(a, b):
    primes = []
    for n in range(a, b + 1):
        if n > 1:
            for i in range(2, int(n//2) + 1):
                if n % i == 0:
                    break
            else:
                primes.append(n)
    return primes

a = int(input("Enter start: "))
b = int(input("Enter end: "))

result = primes_between(a, b)
print(result)
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sum_first_elements(tuples):
    return sum(t[0] for t in tuples)

n = int(input("Enter number of tuples: "))
tuples = []

for _ in range(n):
    t = tuple(map(int, input("|Enter tuple: ").split()))
    tuples.append(t)

print(sum_first_elements(tuples))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

n = int(input("Enter n: "))

print(fib(n))
```

### 09) Write a function to get the name of the student based on the given rollno.
#### Example: Given dict1 = {101:'Ajay', 102:'Rahul', 103:'Jay', 104:'Pooja'} find name of student whose rollno = 103

```python
def get_student(roll, stu):
    if roll in stu:
        return stu[roll]
    else:
        return "Roll number not found"

n = int(input("Enter number of students: "))
students = {}

for _ in range(n):
    r = int(input("Roll: "))
    name = input("Name: ")
    students[r] = name

roll = int(input("Enter roll to search: "))
print(get_student(roll, students))
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sum_0(lists):
    newList = []
    for i in lists:
        if i%10==0:
            newList.append(i)
    return sum(newList)

lists = list(map(int, input("Enter list elements: ").split()))
print(sum_0(lists))
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invert_dict(dict1):
    inv_di = {}
    for k,v in dict1.items():
        inv_di[int(v)] = k
    return inv_di
    
keys = list(input("Enter keys for dict :").split())
values = list(input("Enter value for dict :").split())
dict1 = {k: int(v) for k, v in zip(keys, values)}
inv_di = invert_dict(dict1)
print(inv_di)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def count_case(s):
    uppercase=lowercase=0
    for i in s:
        if i.isupper():
            uppercase += 1
        else:
            lowercase+= 1
    return uppercase , lowercase

s = input("Enter a String :")
uppercase , lowercase = count_case(s)
print("uppercase total :",uppercase)
print("lowercase total :",lowercase)
```
