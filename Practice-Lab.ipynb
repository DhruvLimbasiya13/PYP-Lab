## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
totalStudents = int(input("Enter Total Number of Student :"))
students = []

for i in range(totalStudents):
    name = input("Name: ")
    marks = list(map(int, input("Enter marks: ").split()))
    students.append((i+1 , name , marks))

result = {}

for i in range(totalStudents):
    unique_marks = set(students[i][2])
    roll = i +1 ;
    result[roll] = (students[i][1], round(sum(unique_marks)/len(unique_marks) , 2))

print(students)
print(result)
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
totalitems = int(input("Enter Total Number of items :"))
inventory = {}

for _ in range(totalitems):
    item = input("item name: ")
    stock_price = tuple(map(int, input("Enter price and stock : ").split()))
    inventory[item] = stock_price

totalValue = 0 
for key,value in inventory.items():
    totalValue += value[0]*value[1]
print(inventory)
print("Total Value :",totalValue)
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = list(input("Enter Emails with space separate : ").split())
result = {}
for i in emails:
    em = i.split("@")
    dm = em[1]
    if dm in result.keys() :
        result[dm] += 1
    else:
        result[dm] = 1
print(emails)
print(result)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
totalitems = int(input("Enter Total Number of items in cart:"))
cart = []

for _ in range(totalitems):
    item = input("item name: ")
    unit = int(input("Enter units: "))
    cart.append((item , unit))

print("Now enter the items with price.")
totalitems = int(input("Enter Total Number of items in store for pricing:"))

prices = {}

for _ in range(totalitems):
    item = input("item name: ")
    price = int(input("Enter price: "))
    prices[item] = price
    
totalbill = 0
totalvalue = {}
for i in cart:
    total = prices[i[0]] * i[1]
    totalvalue[i[0]] = total
    totalbill+= total

print(cart)
print(prices)
print(totalvalue)
print("Total Bill :" , totalbill)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
s = input("Enter a String: ")
dict1 = {}
count = 0

for i in s:
    if i in dict1:
        dict1[i].append(count)
    else:
        dict1[i] = [count]
    count += 1

print(dict1)
````
