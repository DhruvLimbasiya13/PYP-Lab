### 04) WAP to print numbers between two given numbers which is divisible by 2 but not divisible by 3.

### 06) WAP to print sum of series 1 + 4 + 9 + 16 + 25 + 36 + ...n.

```python
lastNum = int(input("Enter Last Number :"))
sum = 0 
for i in range(1,lastNum+1):
    sum+=(i*i)
    if i == lastNum:
        print(i*i ,end="=")
    else:
        print(i*i, end="+")
print(sum)
```

### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
lastNum = int(input("Enter Last Number :"))
sum = 0 
for i in range(1,lastNum+1):
    if i == lastNum:
        if i%2!=0:
            sum+=i
        else:
            sum-=i
        print(i ,end="=")
    elif i%2!=0:
        sum+=i
        print(i, end="-")
    else:
        sum-=i
        print(i,end="+")
print(sum)
```

### 10) WAP to print GCD of given two numbers.

```python
num1 = int(input("Enter First Number :"))
num2 = int(input("Enter Second Number :"))
greatest = num1 if num1>=num2 else num2
for i in range(greatest ,0 ,-1):
    if num1%i==0 and num2%i==0:
        print(f"GCD of {num1} and {num2} is {i}")
        break
```

### 12) WAP to find whether the given number is Prime or not.

### 13) WAP to print sum of digits of given number.

```python
num = int(input("Enter Number For Sum of Digits :"))
temp = num
ans=0
while(temp>0):
    ans+=(temp%10)
    temp//=10
print(f"Sum of digit {num} = {ans}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
num = int(input("Enter Number For Sum of Digits :"))
temp = num
ans= rev =0
while(temp>0):
    r=(temp%10)
    rev = rev*10 + r
    temp//=10

if num == rev:
    print(f"{num} is palindrome")
else:
    print(f"{num} is not palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num = int(input("Enter Number For Sum of Digits :"))
temp = num
digits=ans=0
while(temp>0):
    digits+=1
    temp//=10

temp = num
while(temp>0):
    r = temp%10
    ans+=(r**digits)
    temp//=10

if num == ans:
    print(f"{num} is Armstrong")
else:
    print(f"{num} is not Armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
num = int(input("Enter Number For Perfect :"))
sum =0
for i in range(1,num):
    if num%i==0:
        sum+=i
        print(i , end=",")

print()
if sum == num :
    print(f"{num} is Perfect Number.")
else:
    print(f"{num} is Not Perfect Number.")
```
