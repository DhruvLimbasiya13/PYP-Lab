### 04) WAP to find the longest word(s) in a file.

```python
fp=open('new2.txt','r')
text=fp.read()
fp.close()
words=text.split()
max_length=0
for i in words:
    if len(i)>max_length:
        max_length=len(i)
for i in words:
    if len(i)==max_length:
        print(i)
```

### 05) WAP to count the no. of lines, words and characters in a given text file.

```python
fp = open("new2.txt" , "r")

lines = fp.readlines()
fp.close()
print("Total lines :" , len(lines))

word_count = 0
char_count = 0
for i in lines:
    words = i.split()
    word_count  += len(words)
    char_count += len(i)

print("Total words :",word_count)
print("Total characters :",char_count)
```

### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def frequency(search_word):
    fp = open('new2.txt', "r")
    text = fp.read()
    fp.close()

    words = text.split()
    count = 0

    for word in words:
        if word == search_word:
            count += 1

    return count


word = input("Enter the word to search: ")

occurrences = frequency(word)
print("Occurrences of", word, ":", occurrences)
```

### 09) WAP to get the score of five subjects from the user, store them in a file. Fetch those marks and find the highest score.

```python
fp = open("marks.txt", "w")

for i in range(5):
    mark = int(input("Enter marks of subject " + str(i+1) + ": "))
    fp.write(str(mark) + "\n")

fp.close()

fp = open("marks.txt", "r")

marks = []
for line in fp:
    marks.append(int(line.strip()))

fp.close()

highest = marks[0]
for m in marks:
    if m > highest:
        highest = m

print("Highest score:", highest)
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
fp = open("primenumbers.txt", "w")

count = 0
num = 2

while count < 100:
    is_prime = True

    for i in range(2, int(num//2) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        fp.write(str(num) + "\n")
        count += 1

    num += 1

fp.close()
```
