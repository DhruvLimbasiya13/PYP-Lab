### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
t1=tuple(map(int,input(f"Enter a tuple:").split()))
result = []

for i in range(len(t1)):
    num = t1[i]
    if num > 1:
        is_prime = True
        for j in range(2, num):
            if num % j == 0:
                is_prime = False
                break
        if is_prime:
            result.append((i, num))

print(tuple(result))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
rows = int(input("Enter number of rows: "))

T = []

for i in range(rows):
    t1 = tuple(map(int, input(f"Enter tuple {i+1}: ").split()))
    T.append(t1)

T = tuple(T)

cols = len(T[0])

for j in range(cols):
    s = 0
    for i in range(rows):
        s += T[i][j]
    print(s, end=" ")
````
